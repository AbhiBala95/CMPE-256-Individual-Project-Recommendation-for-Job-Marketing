In this section of the codes, the term frequency-inverse document frequency is computed on the keywords in the job and self description columns of the processed user data. Three similarity measures,namely,cosine similarity, manhattan distance and euclidean distance are computed on the resulting TF-IDF vector to obtain a set of similar users for a particular user.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
def findKeywords(example_sent):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(example_sent)
    filtered_sentence = [w for w in word_tokens if not w in stop_words] 
    filtered_sentence = [] 
    for w in word_tokens: 
        if w not in stop_words: 
            filtered_sentence.append(w) 
    return filtered_sentence
```

```python
import nltk
import pandas as pd
import csv
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
df = pd.read_csv('newlinkedindata.csv')
with open('newlinkedindata.csv', 'r', encoding="utf8") as csvfile:
    rows = csv.reader(csvfile)
    next(rows, None)
    newCol = []
    for row in rows:
        sentence = row[19]
        keywords = findKeywords(sentence)
        newCol.append(keywords)
df2 = pd.DataFrame({'Keywords': newCol})
data = pd.concat([user_data, df2], axis=1)
#print(data)
data.to_csv('linkedinkeywords.csv',index=False)
```
