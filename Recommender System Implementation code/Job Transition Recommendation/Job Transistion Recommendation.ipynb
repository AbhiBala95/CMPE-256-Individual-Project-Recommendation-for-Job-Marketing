In this section of codes, the job positions columns in the user dataset are extracted and added into the old and transitioned job columns of the new dataset. The years of experience for each job position is also extracted from the user dataset. The inconsistancy in the years of experience in the form of zero in both duration years and durations months in corrected by assuming a minimum experience of 6 months for all the job positions. Based on the word searched, similar jobs along with suitable job transition is displayed in the ascending order of the year of experience and months of experience
